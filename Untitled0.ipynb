```python
df['Dates'] = pd.to_datetime(df['Dates'])
df['Days'] = (df['Dates'] - df['Dates'].min()).dt.days
```

```python
from scipy.interpolate import interp1d
```

```python
price_model = interp1d(
    df['Days'],
  df['Prices'],
    fill_value='extrapolate'
)
```

```python
def estimate_price(date):

    date = pd.to_datetime(date)

    days = (
        date - df['Dates'].min()
    ).days

    return float(price_model(days))
```

```python
x = np.arange(len(df))

y = df['Prices']

coeffs = np.polyfit(x, y, 1)

slope = coeffs[0]

intercept = coeffs[1]
```

```python
future_dates = pd.date_range(
    start='2024-10-31',
    periods=12,
    freq='M'
)
```

```python
future_prices = []

for i in range(12):

    value = slope * (len(df)+i) + intercept

    future_prices.append(value)
```

```python
future_df = pd.DataFrame({
    'Date': future_dates,
    'Price': future_prices
})

print(future_df.head())
```

```python
import pandas as pd

def price_contract(start_dates, end_dates, strike_prices, quantities, contract_size, risk_free_rate, num_simulations):
    results = []
    for i in range(len(start_dates)):
        start_date = pd.to_datetime(start_dates[i])
        end_date = pd.to_datetime(end_dates[i])
        strike_price = strike_prices[i]
        quantity = quantities[i]

        # Estimate prices using the previously defined estimate_price function
        # Note: estimate_price expects a single date, not an array
        estimated_start_price = estimate_price(start_date)
        estimated_end_price = estimate_price(end_date)

        # Simple calculation for contract value/profit (example)
        # This is a placeholder; a real financial model would be more complex
        contract_profit_per_unit = (estimated_end_price - strike_price) * quantity
        # Discount future value to present for simplicity. A more robust model would use Black-Scholes or similar.
        total_contract_value = contract_profit_per_unit * contract_size / (1 + risk_free_rate)

        results.append(total_contract_value)
    return sum(results) # Return sum of values for all contracts
```

```python
import pandas as pd

def storage_contract_value(
    injection_dates,
    withdrawal_dates,
    purchase_prices,
    sale_prices,
    volume,
    storage_cost_per_month,
    max_storage_volume
):

    if volume > max_storage_volume:
        return "Volume exceeds storage capacity"

    total_value = 0

    for i in range(len(injection_dates)):

        buy_price = purchase_prices[i]
        sell_price = sale_prices[i]

        months_stored = (
            pd.to_datetime(withdrawal_dates[i]).to_period('M')
            -
            pd.to_datetime(injection_dates[i]).to_period('M')
        ).n

        profit = (sell_price - buy_price) * volume

        storage_cost = (
            storage_cost_per_month *
            volume *
            months_stored
        )

        total_value += profit - storage_cost

    return total_value
```

```python
result = storage_contract_value(
    ['2024-01-01'],
    ['2024-06-01'],
    [10],
    [15],
    1000,
    0.1,
    2000
)

print(result)
```
